```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
from sympy import *
from sympy.matrices import Matrix
```

```python
sx=np.matrix([[0,1],[1,0]])
sz=np.matrix([[1,0],[0,-1]])
ide=np.matrix([[1,0],[0,1]])
```

```python
h2,g2,k2=sym.symbols('h2 g2 k2')
h3,g3,k3=sym.symbols('h3 g3 k3')
```

```python
testM = sym.Matrix(h2*sz + g2*sx)
testM
```

```python
Ham2=h2*(np.kron(sz,ide)+np.kron(ide,sz))+g2*(np.kron(sx,ide)+np.kron(ide,sx))-k2*(np.kron(sz,ide)*np.kron(ide,sz))

Ham2=Matrix(Ham2)
Ham2
```

```python
hterm=h3*(np.kron(np.kron(sz,ide),ide)+np.kron(np.kron(ide,sz),ide)+np.kron(np.kron(ide,ide),sz))
gammaterm=g3*(np.kron(np.kron(sx,ide),ide)+np.kron(np.kron(ide,sx),ide)+np.kron(np.kron(ide,ide),sx))
jterm=k3*(
    np.kron(np.kron(sz,ide),ide)*np.kron(np.kron(ide,sz),ide)+
    np.kron(np.kron(ide,sz),ide)*np.kron(np.kron(ide,ide),sz)+
    np.kron(np.kron(sz,ide),ide)*np.kron(np.kron(ide,ide),sz)
)
Ham3=hterm+gammaterm-jterm

Ham3=Matrix(Ham3)
Ham3
```
